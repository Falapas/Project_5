```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
motorcycles = pd.read_excel("Motorcycles.xlsx")
```

```python
motorcycles
```

```python
motorcycles.corr()
```

```python
other_brands = motorcycles[motorcycles.Country != "ITA"]

mean_other_brands = other_brands["Price"].mean()

mean_other_brands
```

```python
ita_brands = motorcycles[motorcycles.Country == "ITA"]

mean_ita_brands = ita_brands["Price"].mean()

ita_brands

#mean_ita_brands
```

```python
# H0: ITA Brand Prices <= Other Brands Price
# H1: ITA Brand Prices > Other Brands Price
# 95% confidence level

stat, pvalue = st.ttest_ind(ita_brands["Price"], other_brands["Price"], equal_var = False)

print("stat:", stat, "p-value:", pvalue/2)
```

```python
# H0: ITA Brand Prices <= 10387
# H1: ITA Brand Prices > 10387
# 95% confidence level

stat, pvalue = st.ttest_1samp(motorcycles[("ITA" == motorcycles["Country"])]["Price"], 10387)

print("stat:", stat, "p-value:", pvalue/2)
```

```python
"""We can not exclude our null hypothesis in both cases, because our p value is higher than 5%, with a 95% condidence level."""
```
